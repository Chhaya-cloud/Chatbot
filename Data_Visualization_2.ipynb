```python
# Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Heatmap -> Colored table
# Creating a data
data = np.random.randint(low = 1, high = 100, size = (10,10))
print('Data to be plotted :\n', data)
```

```python
# High value is never included.
```

```python
# Heatmap
sns.heatmap(data, annot = True, cmap = 'Blues')
```

```python
# Help
help (sns.heatmap)
```

```python
# Heatmap is used typically for correlation analysis in ML module.
```

```python
# catplot -> Categorical scatterplot
# Getting the data
exercise = sns.load_dataset('exercise', index_col = 0 )
exercise.head()
```

```python
# Help
sns.get_dataset_names()
```

```python
# catplot(continued.............)
sns.catplot(x = 'time', y = 'pulse', data = exercise, hue = 'kind')
```

```python
# Hue -> separating the values by that columns
```

```python
# distplot() -> used to density curve along with histogram
plt.figure(figsize = (5, 3))
sns.distplot(exercise['pulse'], kde = False, color = 'green')
```

```python
# distplot() -> used to density curve along with histogram
plt.figure(figsize = (5, 3))
sns.distplot(exercise['pulse'], kde = True, color = 'green')
```

```python
#Distplot with kde = true is density curve plot
#Distplot with kde= flase is simple histogram
```

```python
# loading the 'tips' dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Pairplot
sns.pairplot(df, hue = 'smoker', palette = 'coolwarm')
```

# pairplot is used to get birds eye view of the dataset

# All diagonal elements gives density plot are univariables
# All other plots are combination of bivariables

```python
# Histogram denotes distribution of values
```

```python
# jointplot() -> used ofr analyzing  2 variables along with analyzing a single variable as well.
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
# jointplot() -> used ofr analyzing  2 variables along with analyzing a single variable as well.
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde')

# This graph is used to find density and distribution of the data simultaneously
```

```python
# jointplot() -> used ofr analyzing  2 variables along with analyzing a single variable as well.
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')

# Darker is the color higher is the density in that region and represented with hexagons
```

```python
# jointplot() -> used ofr analyzing  2 variables along with analyzing a single variable as well.
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')

# Histograms along with densityu curve and it plots shadded interval line called regression line
```
