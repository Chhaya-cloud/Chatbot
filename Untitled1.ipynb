```python
# Creating the dataset
df1 = pd.DataFrame ({'Name' : ['Pankaj','SAi','Niraj','Vijay'], 'Age' : [27,28,29,30], 'Address' : ['Nagpur', 'Kanpur', 'SAtara', 'Pune'],
                       'Qualification' : ['Mtech', 'MA', 'MCom', 'M.Hons']}, index = [0,1,2,3])
df2 = pd.DataFrame ({'Name' : ['Rahul','Ayushi','Sharvari','Dhiraj'], 'Age' : [22,23,24,25],
                     'Address' : ['Delhi', 'Wardha', 'Vizag', 'Mumbai'],
                     'Qualification' : ['Mtech', 'MA', 'MCom', 'M.Hons']}, index = [4,5,6,7])
```

```python
# Creating the dataset
df = pd.DataFrame ({'Name' : ['Jane', 'John', 'Ashley', 'Mike', 'Emily', 'Jack', 'Catlin'],
                    'ctg' : ['A', 'A', 'C', 'B', 'B', 'C', 'B'],
                    'val' : np.random.random(7).round(2),
                    'val2' : np.random.randint(1, 10, size=7)})
df.head()
```

```python
# Filtering  according  to Logical operators
df[df['val'] > 0.5]
```

```python
# Logical operator  works on string values
df[df['Name'] > 'Jane']
```

```python
# Alphabetically all the name after Jane would be returned
```

```python
# Adding multiple conditions while filtering
# Adding multiple conditions while filtering  (OR operator)
df[(df['val'] > 0.7) | (df['val2'] == 1)]
```

```python
# Adding multiple conditions while filtering  (AND operator)
df[(df['val'] < 0.5) & (df['val2'] == 1)]
```

```python
# isin() Operator
names = ['Jane', 'Catlin']
df[df['Name'].isin(names)]
```

```python
# Str accessor
df[df['Name'].str.startswith('J')]
```

```python
# Str accessor
df[df['Name'].str.endswith('n')]
```

```python
# Str accessor
df[df['Name'].str.contains('y')]
```

```python
# Tilde Operator
df[~df['Name'].str.endswith('n')]
```

```python
# nlargest
df.nlargest(3, 'val2')
```

```python
# nsmallest
df.nsmallest(3, 'val2')
```

```python
# Creating the dataframe
df = pd.DataFrame({
    'Product' : ['Milk', 'Bread', 'Eggs', 'Butter', 'Cocoa'],
    'Category' : ['Dairy', 'Bakery', 'Dairy', 'Dairy', 'Bakery'],
    'Quantity' : [10, 20, 25, 30, 40],
                  'Amount' : [270, 300, 200, 400, 500]
    })
df.head()
```

```python
# Pivot table with sum aggregation
pivot = df.pivot_table(index = ['Category'], values = ['Amount'], aggfunc = 'sum')
print (pivot)
```

```python
# Pivot table with sum aggregation
pivot = df.pivot_table(index = ['Product', 'Category'], values = ['Amount'], aggfunc = 'sum')
print (pivot)
```

```python
# Pivot table with sum aggregation
pivot = df.pivot_table(index = ['Product', 'Category'], values = ['Amount', 'Quantity'], aggfunc = 'sum')
print (pivot)
```

```python
# Pivot table with sum aggregation
pivot = df.pivot_table(index = ['Product', 'Category'], values = ['Amount', 'Quantity'], aggfunc = ['sum', 'median'])
print (pivot)
```
