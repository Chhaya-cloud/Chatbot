```python
# importing impotant libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Scatter plot
# Creating the data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94,100]
```

```python
# Setting the figure size (in inches)
plt.figure(figsize = (5,3))
```

```python
# Plot
plt.scatter(x, y, color = 'red')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot')
plt.show()
```

```python
# line plot
# creating the data
year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3, 9.9] # corrected syntax for list assignment

# Plot
plt.figure(figsize = (5,3))
plt.plot(year, unemployment_rate, color = 'green', marker = 'o')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('unemployment Rate across years')
plt.show()
```

```python
# Bar Chart
# Creating the data
x = ['C', 'C++', 'Java', 'Python', 'PHP']
y = [22.2, 17.6, 8.8, 8, 7.7]
```

```python
# Setting the figure size
plt.figure(figsize = (5,3))

```

```python
# Plot
plt.bar(x, y, color = 'black', width = 0.6)
plt.xlabel('Programming Languages')
plt.ylabel('Number of students')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
# loading the inbuilt dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Boxplot
# Plot
plt.figure(figsize = (5,3))
sns.boxplot(df['total_bill'])
plt.show()
```

```python
# Histogram for total bill
plt.figure(figsize = (5,3))
sns.histplot(df['total_bill'])
plt.show()
```

```python
# Stacked bar chart
# Creating the data
x = ['A', 'B', 'C', 'D']
y1 = [30, 40, 20, 10]
y2 = [20, 10, 30, 40]
y3 = [10, 20, 40, 30]
y4 = [10, 30, 20, 40]
# Convert the lists to numpy arrays so we can add them correctly
y1, y2, y3, y4 = np.array(y1), np.array(y2), np.array(y3), np.array(y4)
```

```python
# plot bars in stacked manner
plt.figure(figsize = (5,3))
```

```python
# plot bars in stacked manner
plt.bar(x, y1, color = 'green')
plt.bar(x, y2, bottom = y1, color = 'red')
plt.bar(x, y3, bottom = y1 + y2, color = 'blue')
plt.bar(x, y4, bottom = y1 + y2 + y3, color = 'yellow')


# Set x and y values
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Stacked Bar Chart')
plt.legend (['Round 1', 'Round 2', 'Round 3', 'Round 4'])
plt.show()
```

```python
# Pie Chart
# Creating the data
lables = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
# Plot
plt.figure(figsize = (5,3))
plt.pie(sizes, labels = lables, autopct = '%.2f%%', startangle = 120)
plt.title('Pie Chart')
plt.show()
```
