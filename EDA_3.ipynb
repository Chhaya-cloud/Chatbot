```python
'''
Question 5 ->
Perform correlation analysis on the given dataset.
'''
```

```python
# SOlution
df.corr()
```

```python
cm = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True)
plt.show()
```

```python
# Conclusions
#1)'mileage' and 'Year' have a  strong negative correlation
#2) 'mileage' and 'price' have a moderate negative correlation
#3) 'mpg' and 'tax' have a moderate negative correlation
#4) 'Price'and 'mpg' have a moderate negative correlation
#5) 'Price' and 'enginesize'  have a moderate positive correlation
#6) 'Year' and 'Price'  have a moderate positive correlation
```

```python
# Plot the density curve for tax
sns.kdeplot(df['tax'])
```

```python
# data looks bimodal with some outliers in lefgt tail.
```

```python
# Boxplot for 'year' and 'price'
plt.figure(figsize=(20,10))
sns.boxplot(x='year', y='price', data=df, palette='Set3')
plt.show()

```

```python
# Conclusions-> year 1970 seems  to be outlier
# From 1997  till 2020 data seems to be in constant range.
# After 2010,  there has been linear increase in the price  till 2020.
# The number of outliers increases from the year 2014.
```
