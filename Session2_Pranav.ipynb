```python
GLAXO_df = pd.read_csv("GLAXO.csv")
```

```python
GLAXO_df.head()
```

```python
# Filtering the required columns
BEML_df= BEML_df[['Date', 'Close']]
GLAXO_df = GLAXO_df[['Date', 'Close']]
```

```python
# Getting the data types
BEML_df.dtypes
```

```python
GLAXO_df.dtypes
```

```python
# Creating the Index
BEML_df=BEML_df.set_index(pd.DatetimeIndex(BEML_df['Date']))
BEML_df.head()
```

```python
# Creating the Index for GLAXO
GLAXO_df=GLAXO_df.set_index(pd.DatetimeIndex(GLAXO_df['Date']))
GLAXO_df.head()
```

```python
# data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize= (5,3))
plt.plot(GLAXO_df.index, GLAXO_df.Close)
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Closing Price of GLAXO')
plt.show()

```

```python
# Data Visualization
plt.figure(figsize= (5,3))
plt.plot(BEML_df.index, BEML_df.Close)
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Closing Price of BEML')
plt.show()
```

```python
# Adding a 'gain' column
GLAXO_df['gain']=GLAXO_df.Close.pct_change(periods=1)
BEML_df['gain']=BEML_df.Close.pct_change(periods=1)
```

```python
GLAXO_df.head()
```

```python
BEML_df.head()
```

```python
# Dropping the first row witn NAn value
GLAXO_df.dropna(inplace=True)
BEML_df.dropna(inplace=True)
```

```python
# Doubt
BEML_df.head()
GLAXO_df.head()
```

```python
# Plotting the gains
plt.figure(figsize= (5,3))
plt.plot(GLAXO_df.index, GLAXO_df.gain)
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of GLAXO')
plt.show()
```

```python
# Plot the Gain
plt.figure(figsize= (5,3))
plt.plot(BEML_df.index, BEML_df.gain)
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```

```python
# The above chart is not interpretable

# Density plot for gain of Glaxo
import seaborn as sns
plt.figure(figsize= (5,3))
sns.distplot(GLAXO_df['gain'], label ='GLAXO')
plt.xlabel('Gain')
plt.ylabel('Frequency')
plt.title('Density Plot of Gain of Glaxo')
plt.show()
```

```python
# Getting the statistical properties

print('Mean :',round (BEML_df['gain']).mean(),4)
print('Median :',round (BEML_df['gain']).median(),4)
print('Standard Deviation :',round (BEML_df['gain']).std(),4)
```

```python
# Getting the statistical properties

print('Mean :',round (GLAXO_df['gain']).mean(),4)
print('Median :',round (GLAXO_df['gain']).median(),4)
print('Standard Deviation :',round (GLAXO_df['gain']).std(),4)
```

```python
# Getting the 95% of confidence interval
from scipy import stats

GLAXO_df_ci=stats.norm.interval(0.95,loc=GLAXO_df.gain.mean(),scale=GLAXO_df.gain.std())

# COnfidence interval
print('95% confidence interval is :',np.round(GLAXO_df_ci,4))
```

```python
# Getting the 95% of confidence interval
from scipy import stats

BEML_df_ci=stats.norm.interval(0.95,loc=BEML_df.gain.mean(),scale=BEML_df.gain.std())

# COnfidence interval
print('95% confidence interval is :',np.round(BEML_df_ci,4))
```
