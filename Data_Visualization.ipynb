```python
# Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot
# Creating the data
x = [5,7,2,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
# Setting the  figure size (in inches)
plt.figure(figsize=(5,3))
# Plot
plt.scatter(x, y, color = 'red')
# Setting the x and y labels
plt.xlabel('X values')
plt.ylabel('Y values')
# Showing the plot
plt.show()

```

```python
# Line chart
# Creating the data
year = [1920, 1930,1940,1950,1960,1970,1980,1990, 2000,2010,2020]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3, 8.1]
# Plot
plt.figure(figsize=(5,3))
plt.plot(year, unemployment_rate, color = 'green', marker = 'o')
# Giving title to the plot
# Setting the x and y labels
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
# Showing the plot
plt.show()
```

```python
# Bar chart
# Creating the data
data = {
    'C': 20
    ,'C++': 15
    ,'Java': 30
    ,'Python': 35
}
# Setting the  figure size (in inches)
plt.figure(figsize=(5,3))
# Plot
plt.bar(data.keys(), data.values(), color = 'orange')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
# Reading an inbuilt dataset in 'tips'
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Countplot
sns.countplot(x = 'smoker', data = tips)
plt.show()
```

```python
# Pie Chart
# Creating the data
labels = ['C', 'C++', 'Java', 'Python']
sizes = [205, 150, 300, 350]
# Plot
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct='%.2f%%')
plt.title('Pie Chart')
plt.legend()
plt.show()
```

```python
# Heatmap -> Colored table
# Creating the data
data = np.random.randint(low = 1, high = 100,  size = (10,10))
print('Raw data:\n', data)
```

```python
# Heatmap
sns.heatmap(data, annot = True, cmap = 'viridis')
plt.show()
```

```python
# Catplot
exercise = sns.load_dataset('exercise', index_col = 0)
exercise.head()
```

```python
# Categorical plot
sns.catplot(x = 'diet', y = 'pulse', hue = 'kind', data = exercise)
plt.show()
```
