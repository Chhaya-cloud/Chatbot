```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#v Getting the data into dataframe
data = pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
## What is the distribution of sepal lengths and widths among the Iris species?
# Plotting Sepal length and width distribution among species

sns.displot(data, x = 'SepalLengthCm', hue = 'Species', kde = True)
plt.title('Distribution of Sepal Length')
plt.show()

sns.displot(data, x = 'SepalWidthCm', hue = 'Species', kde = True)
plt.title('Distribution of Sepal Width')
plt.show()
```

```python
## Inference: 1 plots displays the distribution of Sepal Length (in cm) across three species of Iris flowers: Iris-setosa, Iris-versicolor, and Iris-virginica.
## Iris-virginica has the largest range of sepal lengths, ranging from 5.5 cm to 7.5 cm
## Iris-Setosa has the smallest ranges of sepal lenghts, ranging between 4.5 cm and 5.8 cm
## Iris-versicolor has the moderate range of sepal lengths, mainly between 5.0 cm and 6.5 cm

## Inferance: 2 plots displays the distribution of Sepal Width (in cm) across three Iris species: Iris-setosa, Iris-versicolor, and Iris-virginica.
## Iris-setosa Has a wider range between 2.5 cm and 4.0 cm, with a peak around 3.4 cm
## Iris-versicolor Has a narrower range between 2.0 cm and 3.4 cm, with a peak around 2.8 cm.
## Iris-virginica distribution overlaps with both Iris-setosa and Iris-versicolor, ranging from approximately 2.2 cm to 3.6 cm, with a peak around 3.0 cm.
```

```python
## 2. How does the petal length vary across different Iris species?
## Plotting petal length across species

sns.lineplot(data = data, x = 'PetalLengthCm', y = 'Species')
plt.title('Petal Length Across Species')
plt.show()


```

```python
sns.histplot(data, x = 'PetalLengthCm', hue = 'Species', kde = True)
plt.title('Distribution of Petal Length')
plt.show()
```

```python
## Inference: plots shows the distribution of Petal Length (in cm) for the three Iris species: Iris-setosa, Iris-versicolor, and Iris-virginica.
## Iris-Setosa petal length is in a small range, between 1.0 cm and 1.9 cm,
## Iris-versicolor petal length ranges from about 3.0 cm to 5.0 cm
## Iris-virginica petal length ranges from about 4.5 cm to 7.0 cm, Iris-setosa can be separated from the other two species based on petal length.
#Iris-versicolor and Iris-virginica overlap slightly in their distributions.
```

```python
## 3. Is there a relationship between sepal length and petal length?
## plotting relationship between sepal length and petal length

sns.scatterplot(data = data, x = 'SepalLengthCm', y = 'PetalLengthCm')
plt.title('Relationship between Sepal Length and Petal Length')
plt.show()
```

```python
## Inference Sepal Length increases, Petal Length tends to increase as well.
```

```python
## 4, How does the petal width vary across different Iris species?
## Plotting petal width across species

sns.boxplot(data = data, y = 'PetalWidthCm', x = 'Species')
plt.title('Petal Width Across Species')
plt.show()
```

```python
## Inference Iris setosa has the smallest petal width (around 0.2-0.3 cm) and is completely separated from the other species
## Iris versicolor is in the middle (around 1.2-1.5 cm)
## Iris virginica has the largest petal width (around 1.8-2.2 cm)
```

```python
## 5. Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?
## Plotting pair plots between features sepal length, sepal width, petal length, petal width

sns.pairplot(data, hue = 'Species')
plt.show()

```

```python
## Inference Strong positive correlation between petal length and petal width
## Sepal width shows weaker correlations with other features
## Clear three-cluster structure in petal-related plots
## More overlap and noise in sepal-related measurements
```

```python
## 6. Can we distinguish between Iris species based on sepal and petal measurements?
## Plotting sepal and petal measurements to distinguish between Iris species

sns.scatterplot(data = data, x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species')
plt.title('Sepal Measurements')
plt.show()
```

```python
## Inference
## Iris setosa in lower left with shorter sepal lengths
## Iris versicolor and Iris virginica occupy middle and right portions and they overlap each other in sepal measurements
## Iris setosa shows a positive correlation between sepal length and width
## versicolor and virginica show a slight negative correlation as sepal length increases, sepal width tends to decrease
```
