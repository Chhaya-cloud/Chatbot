```python
# Getting the data into dataframe
data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
### Question 1: What is the distribution of passenger ages on the Titanic??
# Age Distribution plot

sns.histplot(data['Age'], bins=10, kde=True, color='purple')
plt.title('Age distribution of Titanic passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
## Inference: Most passengers were aged between 20 and 40.
```

```python
## How does the survival rate vary based on passenger class (Pclass)?
# Plotting survival vs Pclass

sns.countplot(x='Pclass', hue='Survived', data=data, palette='Set2')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival vs Passenger Class')
plt.show()
```

```python
## Inference: Passengers travelling in first class has the higher survival rate as compared to 2nd and 3rd class.
```

```python
## Is there a relationship between the fare paid and the likelihood of survival?
## Plotting relation between Ticket fare vs survival

sns.boxplot(x='Survived', y='Fare', data = data, palette='coolwarm')
plt.title('Fare vs Survival')
plt.show()
```

```python
## Inference: Passengers who survived paid comparatively higher fares.
```

```python
## How does the survival rate differ between male and female passengers?
# Plotting Survival rate vs sex

sns.countplot(x='Sex', hue='Survived', data=data, palette='Set2')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival vs Sex')
plt.show()

```

```python
## Inference: Survival rate for female passengers is higher than the mala passengers
```

```python
## What impact does the number of siblings/spouses (SibSp) or parents/children
#(Parch) aboard have on survival?
## Plot survival by Parents or children aboard

sns.countplot(x='Parch', hue='Survived', data=data, palette='coolwarm')
plt.xlabel('Parents or children aboard')
plt.ylabel('Survival rate')
plt.title('Survival vs Parents or children aboard')
plt.show()
```

```python
##Inference: Passengers with 1 or 2 parents/children had higher survival rate.
```

```python
## Are there any significant differences in survival rates among passengers embarked
##from different ports?
## Plot for Survival rate of passengers embarked from different ports

sns.countplot(x='Embarked', hue='Survived', data=data, palette='Set2')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.title('Survival vs Port of Embarkation')
plt.show()
```

```python
## Inference: passengers embarking from port C has higher survival rate.
```
