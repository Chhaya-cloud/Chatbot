```python
#install and then import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn library has the filter parameter named as hue (useful for deeper analysis of the data)
```

```python
#matplotlib to render plots in the notebook
%matplotlib inline # magic function
```

```python
x = [-3, 5, 7]
```

```python
y = [10, 2, 5]
```

```python
fig = plt.figure(figsize=(20,6))

plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-4, 8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.suptitle('Sales Comaprison', size=30, y=1.25)
```

```python
# x1 = [0,1,2,3,4,5]
# y1 = [0,1,4,9,16,25]
x1 = [x for x in range(100)]
y1 = [x**2 for x in x1]
plt.plot(x1,y1)
```
