```python
# Importing the 'BEML.csv'
beml_df = pd.read_csv("C:/Users/jaipu/Documents/Black_Elephant/Weekday_April_25/Statistics/BEML.csv")
beml_df.head()
```

```python
# Making the date column as index
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
# Importing important libraries
import matplotlib.pyplot as plt
import seaborn as sns
plt.plot(glaxo_df.index, glaxo_df.Close)
```

```python
# Question -> Draw the line plot for 'BEML'
plt.plot(beml_df.index, beml_df.Close)
```

```python
# Getting the percentage change on day-to-day basis
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
```

```python
# Getting the idea about the data
beml_df.head()
```

```python
# Getting the idea about the data
glaxo_df.head()
```

```python
# Dropping the NULL values from the dataframe
glaxo_df = glaxo_df.dropna()
beml_df = beml_df.dropna()
```

```python
# Plot for gains
plt.figure(figsize = (7, 5))
plt.plot(glaxo_df.index, glaxo_df.gain)
plt.xlabel('Time')
plt.ylabel('Gain')
plt.show()
```

```python
# Question -> Plot the gains for 'BEML'
plt.figure(figsize = (7, 5))
plt.plot(beml_df.index, beml_df.gain)
plt.xlabel('Time')
plt.ylabel('Gain')
plt.show()
```

```python
# Density Plot
sns.distplot(glaxo_df.gain, label = 'Glaxo')
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Question -> Plot the density plot for 'BEML'
sns.distplot(beml_df.gain, label = 'BEML')
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Statistical Information on GLAXO
print('Mean :', round(glaxo_df.gain.mean(), 5))
print('Standard Deviation :', round(glaxo_df.gain.std(), 5))
```

```python
# Question -> Statistical Information on BEML
print('Mean :', round(beml_df.gain.mean(), 5))
print('Standard Deviation :', round(beml_df.gain.std(), 5))
```

```python
# Importing important libraries
from scipy import stats
```

```python
# Getting the confidence interval
glaxo_df_ci = stats.norm.interval(0.95,
                                 loc = glaxo_df.gain.mean(),
                                 scale = glaxo_df.gain.std())
print('Gain at 95% confidence interval is :', np.round(glaxo_df_ci, 2))
```

```python
# Question -> Get the 95% confidence interval for 'Gain' column of 'BEML'
beml_df_ci = stats.norm.interval(0.95,
                                 loc = beml_df.gain.mean(),
                                 scale = beml_df.gain.std())
print('Gain at 95% confidence interval is :', np.round(beml_df_ci, 2))
```

```python
# What is the probability of making 2% gain or higher in BEML ?
1 - stats.norm.cdf(0.02,
                  loc = beml_df.gain.mean(),
                  scale = beml_df.gain.std())
```

```python
# Question -> What is the probability of making 3% gain or higher in GLAXO ?
1 - stats.norm.cdf(0.03,
                  loc = glaxo_df.gain.mean(),
                  scale = glaxo_df.gain.std())
```
