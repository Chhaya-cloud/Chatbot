```python
# Selecting and filtering data
# Select a specific column
df['Age'].head()
```

```python
# Filter passengers older than 30
df[df['Age']> 30].head()
```

```python
# Groupby and Aggregation
```

```python
# Group by sex and calculate the mean and median age
df.groupby('Sex')['Age'].mean()
```

```python
df.groupby('Sex')['Age'].median()
```

```python
# handle missing values
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing values
df.dropna(subset=['Embarked'],inplace=True)
```

```python
# Data Preprocessing and transformation
```

```python
# Normalize numerical columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
# Apply log transformation
df['Fare_Log'] = np.log1p(df['Fare'])
```

```python
# Convert categorical variables into numerical format using Pandas .get_dummies().
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
# Data Visualization
```

```python
# Basic Plots
```

```python
# histogram of Age
plt.hist(df['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='Survived', y='Age', data=df)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Boxplot of Age by Survival')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
# Select only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Fine-tuning and Optimization
## Convert float64 to float32 to reduce memory usage
df['Fare'] = df['Fare'].astype('float32')
df['Age'] = df['Age'].astype('float32')
print(df.info())
```
