```python
# Creating the dataset
data1 = {'Name' : ['Pranav', 'Sai', 'Niraj', 'Vijay'],
      'Age' : [27,28,29,30],
      'Address' : ['Nagpur', 'Kanpur', 'Satara', 'Pune'],
      'Qualification' : ['ME', 'MBA', 'MSC', 'MCA']
         }

data2 = {'Name' : ['Rahul', 'Ayushi', 'Sharvari', 'Dhiraj'],
      'Age' : [22,23,24,24],
      'Address' : ['Mumbai', 'Delhi', 'Wardha', 'Vizag'],
      'Qualification' : ['ME', 'MBA', 'MSC', 'MCA']
         }
```

```python
# Converting to dataframes
df = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[2,3,6,7])
```

```python
print(df, '\n\n' ,df1)
```

```python
# Concatenation the two given dataframes
res = pd.concat([df, df1])
print(res)
```

```python
# Column concatenation
res = pd.concat([df, df1], axis=1)
print(res)
```

```python
# Axis parameter has only two values 0 & 1
# Default value for axis is 0 -> row wise operation
# If we manually set axis  to 1 -> columnwise operation
```

```python
# Column concatenation
res = pd.concat([df, df1], axis=1 , join ='inner')
print(res)
```

```python
# Column concatenation
res = pd.concat([df, df1], axis=1 , join ='outer')
print(res)
```
